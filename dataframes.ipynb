### 1. Copy the code from the lesson to create a dataframe full of student grades.

```python
# student names
students = ['Sally',
            'Jane',
            'Suzie',
            'Billy',
            'Ada',
            'John',
            'Thomas',
            'Marie',
            'Albert',
            'Richard',
            'Isaac',
            'Alan']

# make students a series
pd.Series(students, name='my_students')

# make column variables and fill with random grades between 60 and 100
math_grades = np.random.randint(
    low=60, high=100, size=len(students))

english_grades = np.random.randint(
    low=60, high=100, size=len(students))

reading_grades = np.random.randint(
    low=60, high=100, size=len(students))

# Create dictionary for DataFrame contents
df_dict = {'name': students,
               'math': math_grades,
               'english': english_grades,
               'reading': reading_grades}

# Create DataFrame with dictionary
df = pd.DataFrame(df_dict)
```

```python
# a. Create a column named passing_english that indicates whether each student has a passing grade in english.
df['passing_english'] = df.english > 70
df

# b. Sort the english grades by the passing_english column. How are duplicates handled?
df.sort_values('passing_english', ascending=True)

# c. Sort the english grades first by passing_english and then by student name. 
# All the students that are failing english should be first, 
# and within the students that are failing english they should be ordered alphabetically. 
# The same should be true for the students passing english. 
# (Hint: you can pass a list to the .sort_values method)
df.sort_values(['passing_english', 'name'], ascending=[True,True])

# d. Sort the english grades first by passing_english, and then by the actual english grade, 
# similar to how we did in the last step.
df.sort_values(['passing_english', 'english'], ascending=[True,True])

# e. Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.
df = df.assign(overall_grade=((df.english+df.math+df.reading)/3))
```

### 2. Load the mpg dataset. Read the documentation for the dataset and use it for the following questions:



```python
# Load mpg dataset
#data('mpg', show_doc=True) # view the documentation for the dataset
mpg = data('mpg') # load the dataset and store it in a variable
mpg

#How many rows and columns are there? ANSWER: 234
mpg.shape

#What are the data types of each column?
mpg.dtypes
    
#Summarize the dataframe with .info and .describe
#print(mpg.info())
#print(mpg.describe())

#Rename the cty column to city.
mpg = mpg.rename(columns={
    'cty':'city'
    })

#Rename the hwy column to highway.
mpg = mpg.rename(columns={
    'hwy':'highway'
    })

#Do any cars have better city mileage than highway mileage? ANSWER: Zero cars have better city mileage.
sum(mpg['city'] > mpg['highway'])

#Create a column named mileage_difference this column should contain the difference 
# between highway and city mileage for each car.
mpg = mpg.assign(mileage_difference=((mpg.highway-mpg.city)))

#Which car (or cars) has the highest mileage difference? ANSWER: Honda Civic & Volkswagen New Beetle
#method 1
mpg.sort_values(['mileage_difference'], ascending=False)
#method 2
mpg[mpg.mileage_difference == mpg.mileage_difference.max()]

#Which compact class car has the lowest highway mileage? The best? 
# ANSWER: Worst - index 220; Best - index 213
mpg.sort_values(
    'highway', ascending=False
)[mpg['class'] == 'compact'].head()

#Create a column named average_mileage that is the mean of the city and highway mileage.
mpg = mpg.assign(
    average_mileage=(mpg.highway+mpg.city)/2)

#Which dodge car has the best average mileage? The worst? ANSWER: Best-index 38; Worst-indexes(55, 70, 60, 66)
mpg.sort_values(
    'average_mileage', ascending=True
)[mpg['manufacturer'] == 'dodge'].head()
```
