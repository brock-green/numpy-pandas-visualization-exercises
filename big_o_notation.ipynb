Curves to graph:\
y = 0n + 1 and label the curve "O(1)"\
y = log(n) and label the curve "O(log n)"\
y = n and label the curve "O(n)"\
y = n âˆ— log(n) and label it "O(n log n)"\
y = ($n^2$) and label it "O(n^2)"\
y = ($2^n$) and label it "O(2^n)"\
y = n! and label it "O(n!)"\
y = ($n^n$) and label it "O(n^n)"

```python
x = (range(1,20))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4 = [(n * math.log(n)) for n in x]
y5 = [n **2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.figure(figsize=(16,8))

plt.plot(x, y1, color ='gray', ls= 'dotted', label = 'y1=$O(1)$', alpha = .2)
plt.plot(x, y2, color ='red', ls= 'dashed', label = 'y2=$O(log n)$', alpha = .3)
plt.plot(x, y3, color ='green', ls= 'dashed', label = 'y3= $O(n)$', alpha = .4)
plt.plot(x, y4, color ='orange', label = 'y4= $O(n log n)$', alpha = .5)
plt.plot(x, y5, color ='lime', label = 'y5= $O(n^2)$', alpha = .6)
plt.plot(x, y6, color ='purple', label = 'y6= $O(2^n)$', alpha = .7)
plt.plot(x, y7, color ='blue', label = 'y7= $O(n!)$', alpha = .8)
plt.plot(x, y8, color ='red', label = 'y8= $O(n^n)$', alpha = .9)

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation', fontsize = 20)
plt.grid(ls= '--')
plt.legend(loc='upper left')

plt.xlim(1,11)
plt.ylim(0,50)

plt.show()
```
