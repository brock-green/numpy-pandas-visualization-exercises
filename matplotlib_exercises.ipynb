```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

### 1. Use matplotlib to plot the following equation:
y = ($x^2$) âˆ’ x + 2

You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.

```python
x = np.array(range(-50,50))
y = x**2 - x+2
plt.plot(x,y, c='green')
plt.annotate('Origin', xy=(0,0), xytext=(0, 1000), arrowprops={})

plt.title('y = $x^2$ - x + 2')
plt.xlabel('x label')
plt.ylabel('y label')

plt.show()
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

$y =\sqrt{x}$ \
y = ($x^3$) \
y = ($2^x$) \
y = 1/(x + 1)

```python
x = range(0,100)
```

```python
y1 =[math.sqrt(n) for n in x]

plt.plot(x,y1)

plt.title('$y=\sqrt{x}$')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
y2 = [n**3 for n in x]


plt.title('$y = x^3$')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.plot(x,y2)
plt.show()
```

```python
y3 = [(2**x) for x in x]

plt.title('$y = 2^x$')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.plot(x, y3)
plt.show()
```

```python
y4 = [(1/(n+1)) for n in x]

plt.title('$y = 1/(x+1)$')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.plot(x,y4)
plt.show()
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.



```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title('$y=\sqrt{x}$')


plt.subplot(222)
plt.plot(x,y2)
plt.title('$y = x^3$')


plt.subplot(223)
plt.plot(x,y3)
plt.title('$y = 2^x$')


plt.subplot(224)
plt.plot(x,y4)
plt.title('$y = 1/(n+1)$')


plt.show()
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.



```python
plt.figure(figsize=(10,8))

plt.title('Different equations')
plt.xlabel('x label')
plt.ylabel('y label')

plt.plot(x, y1, color='gray', label = '$y1=\sqrt{x}$')
plt.plot(x, y2, color='orange', label = '$y2 = x^3$')
plt.plot(x, y3, color='red', label = '$y3 = 2^x$')
plt.plot(x, y4, label='$y4 = 1/(x+1)$')

plt.xlim(0,8)
plt.ylim(0,8)

plt.grid()
plt.legend()

plt.show()
```
